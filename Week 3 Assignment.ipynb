Part 1: Scraping Data from Wikipedia

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# scrape Wikipedia data
df_citydata = pd.read_html(url, header=0)[0]

# display data
df_citydata
```

```python
# remove rows that have Borough as 'Not assigned'
df_citydata1 = df_citydata[df_citydata.Borough != 'Not assigned']
df_citydata1
```

```python
# combine rows with same postcode and order data by Postcode
df_citydata2 = df_citydata1.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(lambda x: ','.join(x.astype(str))).reset_index()

# rename column
#df.rename(columns={"A": "a", "B": "c"})
df_citydata3 = df_citydata2.rename(columns = {'Postcode':'Postal Code'})
df_citydata3
```

```python
# single case for Queen's park where Borough is assigned but not Neighbourhood
df_citydata3.iloc[85,2] = "Queen's Park"
df_citydata3
```

```python
# size of prepared dataframe
df_citydata3.shape
```

Part 2: Adding Latitude and Longitude

```python
# Load location data
df_coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
df_coordinates.head()
```

```python
# Merge df_citydata3 and location data (Note: Since Postal Code was ordered in the same way as the csv file, a single merge function works)
df_citydatacomplete = pd.merge(df_citydata3, df_coordinates, on = 'Postal Code')
df_citydatacomplete
```

Part 3: Clustering

Problem: You are going on a trip to places nearby Toronto, Canada. You don't enjoy walking, but you also want to visit as many cities as possible. You find out that you have enough money to book three hotels in cities nearby Toronto and decide to break up the area around Toronto into three parts, one for each hotel. Where should the hotels be booked so that each hotel is a centralized location of the three parts? For simplicity, assume that the way to minimize walking distance is by having the hotel be located close to the other cities. 

```python
# Import necessary libraries
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
!conda install -c conda-forge folium=0.5.0 --yes
import folium
print('Libraries imported.')
```

```python
from pandas import DataFrame
df_citydataclustering = DataFrame(df_citydatacomplete, columns = ['Latitude', 'Longitude'])
kmeans = KMeans(n_clusters = 3).fit(df_citydataclustering)
centroids = kmeans.cluster_centers_
print(centroids)

# Generate 3 random centroids
plt.scatter(df_citydataclustering['Latitude'], df_citydataclustering['Longitude'], c = kmeans.labels_.astype(float), s= 50, alpha = 0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c = 'red', s = 50)
plt.title('Scatter Plot of Longitude vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
```

```python
# Iterate KMeans for better centroids
k_means = KMeans(init="k-means++", n_clusters = 3, n_init = 12)
k_means.fit(df_citydataclustering)
k_menas_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_

# Display Latitudes and Longitudes of the three central locations
k_means_cluster_centers
```

```python
# Plot with new centroids
plt.scatter(df_citydataclustering['Latitude'], df_citydataclustering['Longitude'], c = kmeans.labels_.astype(float), s= 50, alpha = 0.5)
plt.scatter(k_means_cluster_centers[0,0],k_means_cluster_centers[0,1], c = 'red')
plt.scatter(k_means_cluster_centers[1,0],k_means_cluster_centers[1,1], c = 'red')
plt.scatter(k_means_cluster_centers[2,0],k_means_cluster_centers[2,1], c = 'red')
plt.title('Scatter Plot of Longitude vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
```

```python
# Mapping using Folium (Note: only 3 centroid mapped because mapping every location makes it difficult to read)
# Possible location of hotels
coordinates = pd.DataFrame({'Latitude':[k_means_cluster_centers[0,0],k_means_cluster_centers[1,0], k_means_cluster_centers[2,0]], 
                            'Longitude':[k_means_cluster_centers[0,1],k_means_cluster_centers[1,1], k_means_cluster_centers[2,1]],
                            'Name':['Hotel 1', 'Hotel 2', 'Hotel 3']})
coordinates

# Create empty map
m = folium.Map(location=[43.65, -79.38], zoom_start=10)
 
# Add map markers
folium.Marker([k_means_cluster_centers[0,0], k_means_cluster_centers[0,1]], popup = 'Hotel 1').add_to(m)
folium.Marker([k_means_cluster_centers[1,0], k_means_cluster_centers[1,1]], popup = 'Hotel 2').add_to(m)
folium.Marker([k_means_cluster_centers[2,0], k_means_cluster_centers[2,1]], popup = 'Hotel 3').add_to(m)
m

```
